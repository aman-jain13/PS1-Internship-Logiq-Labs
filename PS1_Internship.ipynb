## **Features**

**The function below returns the locations the carriers ship to (Feature 1)**

```python
def location():
    cities = {}
    for s in slugs:
        cities[s] = set()
    for i in range(0,data.shape[0]):
        y = json.loads(data['checkpoints'][i])
        for j in range(0,len(y)):
            cities[data['slug'][i]].add(y[j]['city'])
    return cities
```

**The function below returns the average shipment time of each carrier (Feature 2)**

```python
def shipment_time():

    shipment_time = {}
    for s in slugs:
        shipment_time[s] = []

    for i in range(0,data.shape[0]):
        start_time = 0
        end_time = 0
        y = json.loads(data["checkpoints"][i])
        for j in range(0,len(y)):

            if y[j]["slug"] == 'dtdc_offline':
                if y[j]["tag"] == 'InfoReceived':
                    start_time = datetime.strptime(y[j]["checkpoint_time"],"%d-%m-%YT%H:%M:%S")
                if y[j]["tag"] == 'Delivered':
                    end_time = datetime.strptime(y[j]["checkpoint_time"],"%d-%m-%YT%H:%M:%S")
            else:
                if y[j]["tag"] == 'InfoReceived':
                    start_time = datetime.fromisoformat(y[j]["checkpoint_time"])
                if y[j]["tag"] == 'Delivered':
                    end_time = datetime.fromisoformat(y[j]["checkpoint_time"])

        if start_time == 0 and end_time!=0:
            for j in range(0,len(y)):
                if y[j]["tag"] == 'PickedUp':
                    start_time = datetime.fromisoformat(y[j]["checkpoint_time"])

        if start_time != 0 and end_time != 0:
            shipment_time[data["slug"][i]].append((end_time-start_time).total_seconds()/86400)

    average_shipment_time = {}
    for s in slugs:
        if len(shipment_time[s]) != 0:
            average_shipment_time[s] = sum(shipment_time[s])/len(shipment_time[s])
    return average_shipment_time
```

**The function below returns the average number of delivery attempts made (Feature 3)**

```python
def delivery_attempts_made():

    number_delivery_attempts = {}
    count = {}
    for s in slugs:
        number_delivery_attempts[s] = 0
        count[s] = 0

    for i in range(0,data.shape[0]):
        if not pd.isna(data['delivery_attempts'][i]):
            number_delivery_attempts[data['slug'][i]] += float(data['delivery_attempts'][i])
            count[data['slug'][i]] += 1
        else:
            y = json.loads(data["checkpoints"][i])
            flag = 0
            for j in range(0,len(y)):
                if y[j]['tag'] == 'OutForDelivery':
                    number_delivery_attempts[data['slug'][i]] += 1.0
                    flag = 1
                if data['slug'][i] == 'xpressbees' and y[j]['message'].lower() == 'out for delivery':
                    number_delivery_attempts[data['slug'][i]] += 1.0
                    flag = 1
            if flag == 1:
                count[data['slug'][i]] += 1
    for s in slugs:
        number_delivery_attempts[s] = number_delivery_attempts[s]/count[s]
    return number_delivery_attempts
```

**The function below returns the average number of orders per day (Feature 4)**


```python
def orders_per_day():

    number_of_orders = {}
    average_orders = {}
    for s in slugs:
        number_of_orders[s] = {}
        average_orders[s] = 0

    for i in range(0,data.shape[0]):
        pickup_date = 0
        y = json.loads(data["checkpoints"][i])
        carrier = data['slug'][i]
        for j in range(0,len(y)):
            if(y[j]['tag'] == 'InfoReceived' or y[j]['tag'] == 'PickedUp'):
                if(y[j]['slug'] == 'dtdc_offline'):
                    pickup_date = datetime.strptime(y[j]["checkpoint_time"],"%d-%m-%YT%H:%M:%S").strftime('%Y/%m/%d')
                else:
                    pickup_date = datetime.fromisoformat(y[j]["checkpoint_time"]).strftime('%Y/%m/%d')
        if pickup_date in number_of_orders[carrier]:
            number_of_orders[carrier][pickup_date] += 1
        else:
            number_of_orders[carrier][pickup_date] = 1

    for s in slugs:
        number_of_orders[s].pop(0)
        average_orders[s] = sum(number_of_orders[s].values())/len(number_of_orders[s])
    return average_orders
```

**The function below returns the number of excpetions of a carrier (Feature 5)**

```python
def exceptions():
    exception = {}
    for s in slugs:
        exception[s] = 0
    for i in range(0, data.shape[0]):
        y = json.loads(data["checkpoints"][i])
        for j in range(0,len(y)):
            if y[j]['tag'] == 'Exception':
                exception[data['slug'][i]] += 1
                break
    return exception
```

**The function below returns the shipment time of carriers on a monthly basis**

```python
def shipment_time_monthly():

    shipment_time = {}
    count = {}
    for s in slugs:
        shipment_time[s] = {}
        count[s] = {}
    for i in range(0,data.shape[0]):
        start_time = 0
        end_time = 0
        month = ""
        y = json.loads(data["checkpoints"][i])
        for j in range(0,len(y)):

            if y[j]["slug"] == 'dtdc_offline':
                if y[j]["tag"] == 'InfoReceived':
                    start_time = datetime.strptime(y[j]["checkpoint_time"],"%d-%m-%YT%H:%M:%S")
                if y[j]["tag"] == 'Delivered':
                    end_time = datetime.strptime(y[j]["checkpoint_time"],"%d-%m-%YT%H:%M:%S")
                    month = end_time.strftime("%m-%Y")
                    if month in count[data['slug'][i]]:
                        count[data['slug'][i]][month] += 1
                    else:
                        count[data['slug'][i]][month] = 1
            else:
                if y[j]["tag"] == 'InfoReceived':
                    start_time = datetime.fromisoformat(y[j]["checkpoint_time"])
                if y[j]["tag"] == 'Delivered':
                    end_time = datetime.fromisoformat(y[j]["checkpoint_time"])
                    month = end_time.strftime("%m-%Y")
                    if month in count[data['slug'][i]]:
                        count[data['slug'][i]][month] += 1
                    else:
                        count[data['slug'][i]][month] = 1

        if start_time == 0 and end_time!=0:
            for j in range(0,len(y)):
                if y[j]["tag"] == 'PickedUp':
                    start_time = datetime.fromisoformat(y[j]["checkpoint_time"])

        if start_time != 0 and end_time != 0:
            if month in shipment_time[data['slug'][i]]:
                shipment_time[data["slug"][i]][month] += (end_time-start_time).total_seconds()/86400
            else:
                shipment_time[data["slug"][i]][month] = (end_time-start_time).total_seconds()/86400

    average_shipment_time = {}
    for s in slugs:
        average_shipment_time[s] = {}
        for month in shipment_time[s].keys():
            average_shipment_time[s][month] = round(shipment_time[s][month]/count[s][month], 2)
    return average_shipment_time
```

**Score calculation**

```python
#Feature 1 - Location
#Feature 2 - Average shipment time
#Feature 3 - Average number of attempts
#Feature 4 - Average number of orders per day (capacity)
#Feature 5 - Number of exceptions

#Normalized features
#(x-xmin)/(xmax-xmin) if higher value indicates higher performance and (xmax-x)/(xmax-xmin) if higher value indicates lower performance
#score = 3*ship_time + 3*locations + 2*daily_orders + 1*delivery_attempts + 1*exceptions

def score_model():
    min_loc = len(locations['dtdc'])
    max_loc = len(locations['dtdc'])
    for s in slugs:
        if min_loc > len(locations[s]):
            min_loc = len(locations[s])
        if max_loc < len(locations[s]):
            max_loc = len(locations[s])

    score = {}
    shipping_time = {}
    loc = {}
    daily_orders = {}
    attempts = {}
    exceptions = {}
    for s in slugs:

        shipping_time[s] = (max(ship_time.values())-ship_time[s])/(max(ship_time.values())-min(ship_time.values()))
        loc[s] = (len(locations[s]) - min_loc)/(max_loc-min_loc)
        daily_orders[s] = (orders_daily[s] - min(orders_daily.values()))/(max(orders_daily.values())-min(orders_daily.values()))
        attempts[s] = (max(delivery_attempts.values())-delivery_attempts[s])/(max(delivery_attempts.values())-min(delivery_attempts.values()))
        exceptions[s] = (max(exception.values())-exception[s])/(max(exception.values())-min(exception.values()))

        score[s] = round(3*shipping_time[s] + 3*loc[s] + 2*daily_orders[s] + 1*attempts[s] + 1*exceptions[s], 2)

    return score
```
